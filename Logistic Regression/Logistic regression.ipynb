```python
train_data.drop(columns = ['Survived' , 'Cabin'], inplace = True)
test_data.drop(columns = ['Cabin'], inplace = True)
```

```python
# Check for missing values
print("Missing values in training data:")
print(train_data.isnull().sum())
print("Missing values in testing data:")
print(test_data.isnull().sum())
```

```python
# Impute missing values
imputer = SimpleImputer(strategy='median')
train_data['Age'] = imputer.fit_transform(train_data[['Age']])
test_data['Age'] = imputer.fit_transform(test_data[['Age']])
```

```python
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

```python
# Check for missing values
print("Missing values in training data:")
print(train_data.isnull().sum())
print("Missing values in testing data:")
print(test_data.isnull().sum())
```

```python
# Encode categorical variables
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
# Drop columns that are not useful for the model
train_data.drop(columns=['Name', 'Ticket'], inplace=True)
test_data.drop(columns=['Name', 'Ticket'], inplace=True)
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
# Separate the target variable from the features
X_train = train_data.drop(columns=['Pclass'])
y_train = train_data['Pclass']

X_test = test_data.drop(columns=['Pclass'])
y_test = test_data['Pclass']
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train,y_train)
```

```python
#Predict for X dataset
y_pred = classifier.predict(X_test)
y_pred
```

```python
len(y_pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy}')
```

```python
print(f'Precision: {precision}')
```

```python
print(f'Recall: {recall}')
```

```python
print('Confusion Matrix:')
print(conf_matrix)
```

```python
print('Classification Report:')
print(class_report)
```

```python
"""Interview Questions:

1. What is the difference between precision and recall?

Precision is the ratio of correctly predicted positive cases to all predicted positive cases.
Recall is the ratio of correctly predicted positive cases to all actual positive cases.
Precision is important when false positives are costly (e.g., spam filters).
Recall is important when missing positives is costly (e.g., disease detection).
Both metrics are especially useful for imbalanced datasets.
Precision and recall often trade off against each other; increasing one can decrease the other.
The F1 score combines both into a single metric for balance"""
```
